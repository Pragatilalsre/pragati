```
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
spark=SparkSession.builder.appName("Spark Dateframe").getOrCreate()
```

```
df=spark.read.options(inferSchema='True' ).csv('/FileStore/tables/movies_noheader.csv')
df.show()
```

```
df1=spark.read.options(inferSchema='True' ).csv('/FileStore/tables/ratings_noheader.csv')
df1.show()
```

```
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, FloatType
```

```
movie_schema=StructType([
    StructField("new_movieId", IntegerType(), True),
     StructField("new_title", StringType(), True),
    StructField("new_genres", StringType(), True)
])
```

```
rating_schema=StructType([
    StructField("new_userId", IntegerType(), True),
    StructField("new_movieId", IntegerType(), True),
     StructField("new_rating", FloatType(), True),
    StructField("new_timestamp", StringType(), True)
])
```

```
df2=spark.read.options(header='False').schema(movie_schema).csv('/FileStore/tables/movies_noheader.csv')
df2.show()
```

```
df2.printSchema()
```

```
df3=spark.read.options(header='False').schema(rating_schema).csv('/FileStore/tables/ratings_noheader.csv')
df3.show()
```

```
df3.printSchema()
```

```
df4=df2
df4=df4.join(df3,'new_movieId', 'inner')
df4.printSchema()
```

```
df4.createOrReplaceTempView('temper')
```

```
spark.sql("select new_movieId as Movie_Id, new_title as Movie_Name, avg(new_rating) as Average_Rating, count(new_rating) as Total_No_of_Ratings from temper group by new_movieId, new_title having Total_No_of_Ratings >=10 order by Average_Rating desc limit 20").show()
```
