```
avg_bonus= df.filter(df.state == "NY").groupBy("state").agg(avg("bonus").alias("avg_bonus")).select("avg_bonus").collect()[0]['avg_bonus']
df.filter((df.state == "NY") & (df.department == "Finance") & (df.bonus > avg_bonus)).show()
```

```
from pyspark.sql.functions import udf
```

```
from pyspark.sql.types import IntegerType
```

```
def salary_increment(salary , age):
    if age > 45:
        return salary+500
    else:
        return salary
    
udf_salary = udf(lambda x,y : salary_increment(x ,y) , IntegerType())
df.withColumn("salary" , udf_salary(col("salary") , col("age"))).show()
```
