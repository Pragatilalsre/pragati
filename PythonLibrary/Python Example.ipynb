```
b1= np.array([1.00000000+0.j, 0.70710678+0.70710678j])
print(b1)
for i in range(len(b1)):
    print("real part" ,b1[i].real)
    print("imaginary part" , b1[i].imag)
```

```
b2 = np.array([1,2,3])
print("Size of the array: ", len(b2))
print("Length of one array element in bytes: ", b2.itemsize)
print("Total bytes consumed by the elements of the array: ", b2.nbytes)
```

```
b3= np.array([ 0,10, 20, 40,60])
b4= np.array([10,30,40])
print(np.intersect1d(b3 , b4))
```

```
b4 = np.array([0, 10, 20, 40, 60, 80])
b5 = np.array([10, 30, 40, 50, 70])
print(np.setdiff1d(b4,b5))
```

```
b6 = np.array([0, 10, 20, 40, 60, 80])
b7 = np.array([10, 30, 40, 50, 70])
print("Unique values that are in only one (not both) of the input arrays:")
print(np.setxor1d(b6,b7))
```

```
b8 = np.array([1, 2])
b9= np.array([4, 5])
print("a > b")
print(np.greater(b8, b9))
print("a >= b")
print(np.greater_equal(b8, b9))
print("a < b")
print(np.less(b8, b9))
print("a <= b")
print(np.less_equal(b8, b9))
```

```
c1 = pd.Series([2, 4, 6, 8, 10])
c2 = pd.Series([1, 3, 5, 7, 9])
c3 = c1 + c2
print("Addition")
print(c3)
print("Subtraction")
c4 = c1 - c2
print(c4)
print("Multiplication")
c5 = c1 * c2
print(c5)
print("Division")
c6 = c1 / c2
print(c6)
```
